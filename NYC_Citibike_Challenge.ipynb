```python
import pandas as pd
import datetime as dt
```

```python
# 1. Create a DataFrame for the 201908-citibike-tripdata data. 
file_to_load = "201908-citibike-tripdata.csv"
citibike_df = pd.read_csv(file_to_load)
citibike_df
```

```python
# 2. Check the datatypes of your columns. 
citibike_df.dtypes
```

```python
# 3. Convert the 'tripduration' column to datetime datatype.
trip_duration = citibike_df.tripduration
trip_duration_datetime = pd.to_datetime(trip_duration, unit='s')
citibike_df['new_trip_duration'] = trip_duration_datetime
```

```python
# 4. Check the datatypes of your columns. 
citibike_df.dtypes
```

```python
# 5. Export the Dataframe as a new CSV file without the index.
citibike_df.to_csv("new_citibike.csv", index = False)
```
